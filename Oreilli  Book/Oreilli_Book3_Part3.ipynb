```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Load the dataset
data = pd.read_csv(
    "/content/drive/MyDrive/Colab Notebooks/OREILLI_ML_DL/Book2_ML_Oreilli/adult.csv", header=None, index_col=False,
    names=['age', 'workclass', 'fnlwgt', 'education', 'education-num',
           'marital-status', 'occupation', 'relationship', 'race', 'gender',
           'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
           'income'])

# Select relevant columns
data = data[['age', 'workclass', 'education', 'gender', 'hours-per-week',
             'occupation', 'income']]

# Convert categorical columns to dummy variables
data_dummies = pd.get_dummies(data)

# Print the columns to verify
print("Columns in data_dummies:\n", data_dummies.columns)

# Adjust column selection based on the actual column names
# Example: Select a subset of columns if you want to include all features
features = data_dummies.drop(columns=['income_<=50K', 'income_>50K'])

# Extract NumPy arrays
X = features.values
y = data_dummies['income_>50K'].values  # Use the correct column for the target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Initialize and train the logistic regression model
logreg = LogisticRegression(max_iter=1000)  # Increase max_iter if needed
logreg.fit(X_train, y_train)

# Print the test score
print("Test score: {:.2f}".format(logreg.score(X_test, y_test)))
```

```python
import mglearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

# Load the data
X, y = mglearn.datasets.make_wave(n_samples=100)

# Define bin edges
bins = np.linspace(-3, 3, 6)  # 5 bins

# Create the binning encoder
encoder = KBinsDiscretizer(n_bins=len(bins)-1, encode='onehot', strategy='uniform')

# Transform the original data
X_binned = encoder.fit_transform(X.reshape(-1, 1))

# Combine the original and binned features
X_combined = np.hstack([X, X_binned.toarray()])

# Fit the LinearRegression model on combined data
reg = LinearRegression().fit(X_combined, y)

# Create a line of values to predict
line = np.linspace(-3, 3, 1000).reshape(-1, 1)

# Transform the line data and combine with original line data
line_binned = encoder.transform(line)
line_combined = np.hstack([line, line_binned.toarray()])

# Make predictions
predictions = reg.predict(line_combined)

# Plot the predictions from Linear Regression
plt.plot(line, predictions, label='Linear Regression Combined')

# Plot vertical lines for the bins
for bin in bins:
    plt.plot([bin, bin], [-3, 3], ':', c='k')

# Plot the original data points
plt.plot(X[:, 0], y, 'o', c='k', label='Data Points')

# Add labels and legend
plt.ylabel("Regression Output")
plt.xlabel("Input Feature")
plt.legend(loc="best")

# Show the plot
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler

# Load the dataset
housing = fetch_california_housing()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    housing.data, housing.target, random_state=0
)

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform it
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test data
X_test_scaled = scaler.transform(X_test)

# Initialize PolynomialFeatures with degree 2
poly = PolynomialFeatures(degree=2)

# Fit and transform the training data
X_train_poly = poly.fit_transform(X_train_scaled)

# Transform the test data
X_test_poly = poly.transform(X_test_scaled)

# Print the shapes
print("X_train.shape: {}".format(X_train.shape))
print("X_train_poly.shape: {}".format(X_train_poly.shape))
print("X_test.shape: {}".format(X_test.shape))
print("X_test_poly.shape: {}".format(X_test_poly.shape))
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train_scaled, y_train)
print("Score without interactions: {:.3f}".format(
 ridge.score(X_test_scaled, y_test)))
ridge = Ridge().fit(X_train_poly, y_train)
print("Score with interactions: {:.3f}".format(
 ridge.score(X_test_poly, y_test)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100).fit(X_train_scaled, y_train)
print("Score without interactions: {:.3f}".format(
    rf.score(X_test_scaled, y_test)))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the dataset (example using Iris dataset)
data = load_iris()
X = data.data
y = data.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0
)

# Initialize the scaler and standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Example of feature selection: Select top 2 features based on ANOVA F-value
select = SelectKBest(f_classif, k=2)
X_train_selected = select.fit_transform(X_train_scaled, y_train)
X_test_selected = select.transform(X_test_scaled)

# Fit and evaluate Logistic Regression with all features
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_scaled, y_train)
print("Score with all features: {:.3f}".format(lr.score(X_test_scaled, y_test)))

# Fit and evaluate Logistic Regression with selected features
lr.fit(X_train_selected, y_train)
print("Score with only selected features: {:.3f}".format(
    lr.score(X_test_selected, y_test)))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example DataFrame creation
dates = pd.date_range(start='2023-01-01', periods=200, freq='D')
rentals = np.random.randint(1, 100, size=(200,))
citibike = pd.DataFrame(data=rentals, index=dates, columns=['rentals'])

# Extract the target values (number of rentals)
y = citibike['rentals'].values

# Convert the time index to POSIX time (seconds since epoch)
X = citibike.index.astype(int) // 10**9  # Convert nanoseconds to seconds
X = X.values.reshape(-1, 1)  # Reshape to a 2D array for modeling

# Define the number of training samples
n_train = 184

# Function to evaluate and plot a regressor on a given feature set
def eval_on_features(features, target, regressor):
    n_train = 184
    X_train, X_test = features[:n_train], features[n_train:]
    y_train, y_test = target[:n_train], target[n_train:]

    regressor.fit(X_train, y_train)

    print("Test-set R^2: {:.2f}".format(regressor.score(X_test, y_test)))

    y_pred_train = regressor.predict(X_train)
    y_pred_test = regressor.predict(X_test)

    plt.figure(figsize=(12, 6))

    plt.plot(range(n_train), y_train, label="Train Actual", color='blue')
    plt.plot(range(n_train, len(y_test) + n_train), y_test, label="Test Actual", color='green')
    plt.plot(range(n_train), y_pred_train, '--', label="Train Prediction", color='blue')
    plt.plot(range(n_train, len(y_test) + n_train), y_pred_test, '--', label="Test Prediction", color='green')

    plt.xlabel("Index")
    plt.ylabel("Rentals")
    plt.legend(loc='best')
    plt.title("Actual vs Predicted Rentals")
    plt.show()

# Create and evaluate a regressor
regressor = LinearRegression()
eval_on_features(X, y, regressor)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()

# Split data into train+validation set and test set
X_trainval, X_test, y_trainval, y_test = train_test_split(
    iris.data, iris.target, random_state=0
)

# Initialize variables to keep track of the best score and parameters
best_score = 0
best_parameters = {}

# Perform grid search with cross-validation
for gamma in [0.001, 0.01, 0.1, 1, 10, 100]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        # Create an SVC model with current parameters
        svm = SVC(gamma=gamma, C=C)

        # Perform cross-validation
        scores = cross_val_score(svm, X_trainval, y_trainval, cv=5)

        # Compute mean cross-validation accuracy
        score = np.mean(scores)

        # If we got a better score, store the score and parameters
        if score > best_score:
            best_score = score
            best_parameters = {'C': C, 'gamma': gamma}

# Rebuild a model on the combined training and validation set
svm = SVC(**best_parameters)
svm.fit(X_trainval, y_trainval)

# Optionally evaluate on the test set
test_score = svm.score(X_test, y_test)
print("Best cross-validation score: {:.2f}".format(best_score))
print("Best parameters: ", best_parameters)
print("Test set score with best parameters: {:.2f}".format(test_score))
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split data into train+validation set and test set
X_trainval, X_test, y_trainval, y_test = train_test_split(
    X, y, random_state=0
)

# Define parameter grid
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'gamma': [0.001, 0.01, 0.1, 1, 10, 100]
}

print("Parameter grid:\n{}".format(param_grid))

# Initialize SVC
svm = SVC()

# Initialize GridSearchCV with the parameter grid
grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy')

# Perform grid search
grid_search.fit(X_trainval, y_trainval)

# Get the best parameters and score
best_parameters = grid_search.best_params_
best_score = grid_search.best_score_

print("Best cross-validation score: {:.2f}".format(best_score))
print("Best parameters: {}".format(best_parameters))

# Rebuild the model with the best parameters and evaluate on the test set
best_svm = SVC(**best_parameters)
best_svm.fit(X_trainval, y_trainval)
test_score = best_svm.score(X_test, y_test)

print("Test set score with best parameters: {:.2f}".format(test_score))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.datasets import load_iris
import mglearn

# Load dataset and split it
iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=0
)

# Define parameter grids
param_grid_linear = {
    'C': np.linspace(1, 2, 6),
    'gamma': np.linspace(1, 2, 6)
}
param_grid_one_log = {
    'C': np.linspace(1, 2, 6),
    'gamma': np.logspace(-3, 2, 6)
}
param_grid_range = {
    'C': np.logspace(-3, 2, 6),
    'gamma': np.logspace(-7, -2, 6)
}

# Plot setup
fig, axes = plt.subplots(1, 3, figsize=(13, 5))

# Grid search and plotting for each parameter grid
for param_grid, ax in zip([param_grid_linear, param_grid_one_log, param_grid_range], axes):
    grid_search = GridSearchCV(SVC(), param_grid, cv=5)
    grid_search.fit(X_train, y_train)

    # Extract mean test scores and reshape them
    scores = grid_search.cv_results_['mean_test_score']
    scores = scores.reshape(len(param_grid['C']), len(param_grid['gamma']))

    # Plot the mean cross-validation scores
    scores_image = mglearn.tools.heatmap(
        scores, xlabel='gamma', ylabel='C',
        xticklabels=param_grid['gamma'], yticklabels=param_grid['C'],
        cmap="viridis", ax=ax
    )
    ax.set_title(f'Grid Search Results\n{list(param_grid.keys())[0]} vs {list(param_grid.keys())[1]}')

# Add colorbar to the plots
plt.colorbar(scores_image, ax=axes.tolist())

# Show the plots
plt.tight_layout()
plt.show()
```
